```python
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
students_performance = pd.read_csv('data/StudentsPerformance.csv')
students_performance.columns = students_performance.columns.str.replace(' ', '_')
students_performance.head()
```

```python
# Прямо из pandas можно вызывать визуализацию:
students_performance.math_score.hist()
```

```python
students_performance.plot.scatter(x='math_score', y='reading_score')
# Отмечаем довольно линейную взаимосвязь
```

```python
# В надстройке seaborn более локаничный интерфейс.
sns.lmplot(x='math_score', y='reading_score', data=students_performance)
# Видим на точках расположенная регрессионная прямая которая показывает ленейную апроксимацию взаимосвязи двух переменных.
```

```python
# Добавив gender можем увидеть как распределяются точки по полам. hue(оттенок) - показывает какой столбце используется для определения цвета.
sns.lmplot(x='math_score', y='reading_score', hue='gender', data=students_performance)
```

```python
# через fit_reg=False можно отключить регрессионные прямые
sns.lmplot(x='math_score', y='reading_score', hue='gender', data=students_performance, fit_reg=False)
```

```python
# Можно установить свои названия осей
ax = sns.lmplot(x='math_score', y='reading_score', hue='gender', data=students_performance, fit_reg=False)
ax.set_xlabels('Math score')
ax.set_ylabels('Reading score')
```
